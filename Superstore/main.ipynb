### c. How do monthly or yearly sales trends look over time?

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales=df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
monthly_trends=monthly_sales.groupby(['Year', 'Month'])['Sales'].sum().unstack(level=0)
monthly_trends
# monthly_sales
```

```python
yearly_trends = df.groupby('Year')['Sales'].sum()
yearly_trends
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
sns.lineplot(monthly_trends)
plt.title('Monthly Sales Trends (All Years)', fontsize=14)
plt.ylabel('Total Sales', fontsize=12)
plt.legend(title='Year', fontsize=10)
plt.grid(True)

plt.subplot(1,2,2)
sns.barplot(x=yearly_trends.index, y=yearly_trends.values, hue=yearly_trends.index)
plt.title('Yearly Sales Trends', fontsize=14)
plt.ylabel('Total Sales', fontsize=12)
plt.grid()

plt.ylabel('Sales')
plt.grid()
plt.show()
```

### d. What are the peak and low sales months for each year?

```python
peak_low_sales = monthly_sales.groupby('Year').apply(lambda group: pd.DataFrame({'Peak Month': [group.loc[group['Sales'].idxmax(), 'Month']],'Low Month': [group.loc[group['Sales'].idxmin(), 'Month']]})).reset_index(level=1, drop=True)
print(peak_low_sales)
```

```python
plt.plot(peak_low_sales, marker=('x'))
plt.show()
```

# 4. Sales Performance by Region

```python
df.rename({'Country/Region': 'Region'}, axis=1).head(2)
```

### a. Which region has the highest sales, and how do sales compare across regions?

```python
region=df.groupby(['Region'])['Sales'].sum().sort_index()
plt.pie(region, labels=region.index, autopct='%0.2f%%', colors=sns.color_palette('pastel'), explode=[0.21,0.041,0.1, 0.02],  counterclock=False, textprops={'fontsize':14})
sns.barplot(region,width=0.5)
plt.title("Distribution of Sales across Regions")
plt.show()
```

### b. What is the distribution of sales across different states and cities?

```python
State_sales=df.groupby(['State'])['Sales'].sum().sort_values(ascending=False)
State_sales.head()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x=State_sales.index,y=State_sales.values, hue=State_sales.index)
plt.xticks(rotation=90,fontsize=11)
plt.title("Top States with Hight Sales")
plt.show()
```

```python
City_sales=df.groupby(['City'])['Sales'].sum().sort_values(ascending=False)[:20]
City_sales.head()
```

```python
sns.barplot(y=City_sales.index,x=City_sales.values, hue=City_sales.index)
plt.title("Top 20 Cities with Hight Sales")
plt.show()
```
